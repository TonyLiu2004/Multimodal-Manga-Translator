```python
from googletrans import Translator
```

```python
async def translate_manga_panel(text):
    # Initialize the translator
    async with Translator() as translator:
        ocr_lines = text
        
        # Translate the batch
        # src='ja' for Japanese, dest='en' for English
        results = await translator.translate(ocr_lines, src='ja', dest='en')
        
        for result in results:
            print(f"Original: {result.origin}")
            print(f"English: {result.text}")
            print("-" * 20)
```

```python
lines = ["お！イルカか．．．", "影分身の術！"]
translate_manga_panel(lines)
```
