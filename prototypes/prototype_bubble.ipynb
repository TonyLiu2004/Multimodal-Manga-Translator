```python
from ultralytics import YOLO
from PIL import Image, ImageFont
import sys
import os
from pathlib import Path
```

```python
parent_dir = str(Path(os.getcwd()).parent)

if parent_dir not in sys.path:
    sys.path.append(parent_dir)

base = Path.cwd().parent
```

```python
model_path = base / "backend/models/kiuyha.pt"
model = YOLO(model_path)

def predict(img_path, conf=0.2, iou=0.4, show_labels=True, show_conf=True, imgsz=640):
    img = Image.open(img_path)
    results = model.predict(
        source=img,
        conf=conf,
        iou=iou,
        show_labels=show_labels,
        show_conf=show_conf,
        imgsz=imgsz,
    )
    return results[0]
    
```

```python
image_path = base / "test_images/test_1.png"
img = predict(image_path)
img.show()
```
