```python
# tokenizer = AutoTokenizer.from_pretrained("./GlmOcr/tokenizer")
processor = AutoProcessor.from_pretrained("./GlmOcr/processor")
model = AutoModelForImageTextToText.from_pretrained("./GlmOcr/model")
```

```python
messages = [
    {
        "role": "user",
        "content": [
            {
                "type": "image",
                "url": "naruto_1.png"
            },
            {
                "type": "text",
                "text": "Text Recognition:"
            }
        ],
    }
]

inputs = processor.apply_chat_template(
    messages,
    tokenize=True,
    add_generation_prompt=True,
    return_dict=True,
    return_tensors="pt"
).to(model.device)
inputs.pop("token_type_ids", None)
generated_ids = model.generate(**inputs, max_new_tokens=8192)
output_text = processor.decode(generated_ids[0][inputs["input_ids"].shape[1]:], skip_special_tokens=True)
print(output_text)
```
